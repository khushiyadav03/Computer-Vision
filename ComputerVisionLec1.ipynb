```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/painted-natural-mountain-scenery-with-colorful-flowers-free-image.webp")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR se RGB me convert becoz in open cv by default color format is BGR
plt.imshow(img)
plt.title("Step 1: RGB Image")
plt.axis("off")
plt.show()
```

```python
print(img.shape) # shape of the img : height * width && 3-> no of color chanels if(3 then color img else black nd white)
```

```python
print(img[0:5, 0:5])
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
plt.title("Step 2: Gray Image")
plt.axis("off")
plt.show()
```

```python
print(gray[0:10, 0:10])
```

```python
sampled = cv2.resize(gray, (50, 50))
plt.imshow(sampled, cmap="gray")
plt.title("Step 3: Sampled Image (Lower Resolution)")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(sampled, cmap="gray")
plt.grid(True)
plt.title("Pixel Grid")
plt.axis("off")
plt.show()
```

**resolution** -> Resolution refers to how many pixels are used to represent an image

**sampling** -> process of converting a continuous real-world scene into discrete pixels. types: downsampling and upsampling

**Downsampling** means reducing the number of pixels in an image or feature map

**Upsampling** means increasing the number of pixels in an image or feature map

A **photodiode** is a p–n junction semiconductor that generates current when light hits it.


In computer vision cameras:

 Every pixel = one photodiode


Millions of photodiodes form the image sensor.

# Bayer Filter
The Bayer filter is a color filter array (CFA) placed on top of an image sensor in most digital cameras. It allow the sensor to capture color images using sensors that can only measure light intensity.


**Bayer Pattern Layout**

The Bayer filter uses a repeating 2 × 2 pattern:

G   R

B   G

This pattern is called RGGB.

#Quantization
Quantization is the process of converting continuous intensity values into discrete digital levels so a computer can store and process an image.

A camera sensor measures light as analog electrical signals. Computers cannot store analog values directly, so they must be rounded into fixed numerical levels. That rounding step is quantization.

```python
quantized = (gray // 64) * 64
plt.imshow(quantized, cmap="gray")
plt.title("Step 4: Quantized Image")
plt.axis("off")
plt.show();
```
