**Gaussian Noise** -> andom pixel variation that follows a normal (Gaussian) distribution

That means:

*   Most noise values are small
*   Extreme changes are rare
*   It spreads like a smooth bell curve

**Salt & Pepper Noise** -> random pixels turned completely black (0) or white (255)

* “Salt” → white pixels
* “Pepper” → black pixels
* Everything else stays unchanged

It’s impulse noise: sudden, extreme corruption

**Poisson Noise** -> noise caused by random arrival of photons

* Happens in low-light imaging
* Strongly tied to signal intensity
* More photons → relatively less noise
* Fewer photons → chaos increases

It’s not hardware failure. It’s nature being probabilistic
