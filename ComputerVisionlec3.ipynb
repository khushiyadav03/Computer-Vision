```python
from tensorflow.keras.datasets import mnist
import numpy as np
import cv2
import matplotlib.pyplot as plt

# load dataset
(X_train, _), (_, _) = mnist.load_data()

# pick one image
img = X_train[0]

# plot image
plt.imshow(img, cmap='gray')
plt.title("Original MNIST Image")
plt.axis('off')
plt.show()
```

```python
noise = np.random.normal(0, 30, img.shape)
noisy_gaussian = np.clip(img + noise, 0, 255).astype(np.uint8)
```

```python
sp_img = img.copy()
prob = 0.1

rand = np.random.rand(*img.shape)
sp_img[rand < prob] = 0
sp_img[rand > 1 - prob] = 255
```

```python
mean_img = cv2.blur(noisy_gaussian, (3, 3))
```

```python
meadian_img = cv2.medianBlur(sp_img, 3)
```

```python
plt.figure(figsize=(8, 8))

plt.subplot(2, 2, 1)
plt.imshow(noisy_gaussian, cmap='gray')
plt.title("Gaussion Noise")

plt.subplot(2, 2, 2)
plt.imshow(mean_img, cmap='gray')
plt.title("Salt and Pepper Noise")

plt.subplot(2, 2, 3)
plt.imshow(sp_img, cmap='gray')
plt.title("Mean Filter")

plt.subplot(2, 2, 4)
plt.imshow(meadian_img, cmap='gray')
plt.title("Median Filter")

plt.show()
```
