1. **Spatial Domain (Pixel World)** : work directly on pixels.

Pros:

* Easy to understand
* Fast for simple operations
* Direct manipulation

Cons:

* Hard to isolate patterns like noise vs edges
* Not ideal for complex filtering

2. **Frequency Domain (Pattern World)** : convert the image into frequency components using:
* Fourier Transform (FT)
* Fast Fourier Transform (FFT)

Now instead of pixels, you deal with:
* Low frequencies → smooth areas
* High frequencies → edges, noise

Pros:

* Powerful for noise removal
* Better for global patterns
* Efficient for complex filters

Cons:

* Harder to understand
* More computation
* Not intuitive



**Low-Pass Filter:** A low-pass filter allows low frequencies to pass and blocks high frequencies.

* Low frequency → smooth areas (background, gradual changes)
* High frequency → edges, details, noise

**High-Pass Filter:** A high-pass filter allows high frequencies to pass and blocks low frequencies.

* High frequency → edges, fine details, noise
* Low frequency → smooth background
