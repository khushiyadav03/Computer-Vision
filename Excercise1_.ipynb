```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/painted-natural-mountain-scenery-with-colorful-flowers-free-image.webp")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR se RGB me convert becoz in open cv by default color format is BGR
plt.imshow(img)
plt.title("Step 1: RGB Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap="gray")
plt.title("Gray Image")
plt.axis("off")
plt.show()
```

```python
# lower sampling of gray scale image -> reducing resolution
lower_gray = cv2.resize(gray, (int(gray.shape[1]/2), int(gray.shape[0]/2)), interpolation=cv2.INTER_AREA)
plt.imshow(lower_gray, cmap="gray")
plt.title("Lower Gray Image")
plt.axis("off")
plt.show()
```

```python
# quantization on grayscale img
num_levels = 16 # Example: 16 intensity levels (4 bits per pixel)

# Calculate the step size for quantization
step_size = 256 / num_levels

# Apply quantization
quantized_gray = (gray // step_size) * step_size

plt.imshow(quantized_gray, cmap="gray")
plt.title(f"Quantized Gray Image ({num_levels} levels)")
plt.axis("off")
plt.show()
```
