```python
import torch
import torchvision
import torchvision.transforms as transforms
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# ---------------------------
# 1. Load MNIST test set
# ---------------------------
transform = transforms.ToTensor()

test_dataset = torchvision.datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=1, shuffle=True)
```

```python
# ---------------------------
# 2. Load pretrained CNN
# (Assume you already trained and saved it)
# ---------------------------
class CNN(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = torch.nn.Conv2d(1, 32, 3)
        self.conv2 = torch.nn.Conv2d(32, 64, 3)
        # Corrected input dimension for fc1
        self.fc1 = torch.nn.Linear(36864, 128)
        self.fc2 = torch.nn.Linear(128, 10)

    def forward(self, x):
        x = torch.relu(self.conv1(x))
        x = torch.relu(self.conv2(x))
        x = torch.flatten(x, 1)
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

model = CNN()

try:
    model.load_state_dict(torch.load("mnist_cnn.pth"))
    print("Pre-trained model loaded successfully.")
except FileNotFoundError:
    print("Pre-trained model 'mnist_cnn.pth' not found. Training a new model...")

    # Load MNIST training set (assuming transforms is already defined from cell 1)
    train_dataset = torchvision.datasets.MNIST(
        root='./data',
        train=True,
        download=True,
        transform=transforms.ToTensor()
    )
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=64, shuffle=True)

    # Define loss function and optimizer
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    # Training loop
    num_epochs = 1 # A small number of epochs for demonstration
    for epoch in range(num_epochs):
        for i, (images, labels) in enumerate(train_loader):
            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Backward and optimize
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            if (i+1) % 100 == 0:
                print(f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{len(train_loader)}], Loss: {loss.item():.4f}')

    # Save the trained model
    torch.save(model.state_dict(), "mnist_cnn.pth")
    print("New model trained and saved as 'mnist_cnn.pth'.")

model.eval()
```

```python
# ---------------------------
# 3. Noise Functions
# ---------------------------

# Gaussian Noise N(0, 0.2)
def add_gaussian_noise(image):
    noise = np.random.normal(0, 0.2, image.shape)
    noisy = image + noise
    return np.clip(noisy, 0, 1)

# Salt & Pepper Noise (5%)
def add_salt_pepper(image, prob=0.05):
    noisy = image.copy()
    rand = np.random.rand(*image.shape)

    noisy[rand < prob/2] = 0      # pepper
    noisy[rand > 1 - prob/2] = 1  # salt

    return noisy
```

```python
# ---------------------------
# 4. Denoising Filters
# ---------------------------

# Mean Filter (3x3)
def mean_filter(image):
    return cv2.blur(image, (3,3))

# Median Filter (3x3)
def median_filter(image):
    img_uint8 = (image * 255).astype(np.uint8)
    filtered = cv2.medianBlur(img_uint8, 3)
    return filtered / 255.0
```

```python
# ---------------------------
# 5. Run on one sample
# ---------------------------
images, labels = next(iter(test_loader))

img = images[0].numpy().squeeze()

# Add noise
gaussian_img = add_gaussian_noise(img)
sp_img = add_salt_pepper(img)

# Apply filters
gaussian_denoised = mean_filter(gaussian_img)
sp_denoised = median_filter(sp_img)
```

```python
# ---------------------------
# 6. Visualization
# ---------------------------
titles = ["Original", "Gaussian Noise", "Mean Filter",
          "Salt & Pepper", "Median Filter"]

imgs = [img, gaussian_img, gaussian_denoised,
        sp_img, sp_denoised]

plt.figure(figsize=(10,4))
for i in range(5):
    plt.subplot(1,5,i+1)
    plt.imshow(imgs[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```

```python
# ---------------------------
# 7. Model Predictions
# ---------------------------
def predict(image):
    tensor = torch.tensor(image).unsqueeze(0).unsqueeze(0).float()
    with torch.no_grad():
        output = model(tensor)
        return torch.argmax(output, dim=1).item()

print("Original:", predict(img))
print("Gaussian:", predict(gaussian_img))
print("Mean Filtered:", predict(gaussian_denoised))
print("Salt & Pepper:", predict(sp_img))
print("Median Filtered:", predict(sp_denoised))
```

```python
# result table
results_summary = {
    "Original Clean MNIST Data": predict(img),
    "Gaussian Noise (0.2)": predict(gaussian_img),
    "Salt & Pepper (5% density)": predict(sp_img),
    "Gaussian + Mean Filter": predict(gaussian_denoised),
    "Salt & Pepper + Median Filter": predict(sp_denoised)
}

for key, value in results_summary.items():
    print(f"{key}: {value}")
```

```python
# ---------------------------
# 8. Separate Accuracy Calculation
# ---------------------------

def calculate_accuracy_with_transform(model, dataloader, transform_func=None):
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in dataloader:
            processed_images = []
            for img_tensor in images:
                # Convert torch tensor (1, 28, 28) to numpy array (28, 28)
                img_np = img_tensor.squeeze().numpy()
                if transform_func:
                    transformed_img_np = transform_func(img_np)
                else:
                    transformed_img_np = img_np
                # Convert numpy array (28, 28) back to torch tensor (1, 28, 28)
                processed_images.append(torch.tensor(transformed_img_np).unsqueeze(0).float())

            # Stack processed images back into a batch tensor
            if processed_images:
                batch_processed_images = torch.stack(processed_images)
            else:
                batch_processed_images = images.float() # Fallback for no transforms or empty batch

            outputs = model(batch_processed_images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    return 100 * correct / total

# Helper functions for combined transforms
def gaussian_noise_only(image_np):
    return add_gaussian_noise(image_np)

def salt_pepper_only(image_np):
    return add_salt_pepper(image_np)

def gaussian_mean_filtered(image_np):
    noisy_img = add_gaussian_noise(image_np)
    return mean_filter(noisy_img)

def salt_pepper_median_filtered(image_np):
    noisy_img = add_salt_pepper(image_np)
    return median_filter(noisy_img)


print(f"Accuracy for Original Data: {calculate_accuracy_with_transform(model, test_loader):.2f}%")
print(f"Accuracy for Gaussian Noise: {calculate_accuracy_with_transform(model, test_loader, gaussian_noise_only):.2f}%")
print(f"Accuracy for Salt & Pepper Noise: {calculate_accuracy_with_transform(model, test_loader, salt_pepper_only):.2f}%")
print(f"Accuracy for Gaussian Noise + Mean Filter: {calculate_accuracy_with_transform(model, test_loader, gaussian_mean_filtered):.2f}%")
print(f"Accuracy for Salt & Pepper Noise + Median Filter: {calculate_accuracy_with_transform(model, test_loader, salt_pepper_median_filtered):.2f}%")
```

```python
# ---------------------------
# 9. Calculate Overall Accuracy
# ---------------------------
correct = 0
total = 0
with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f'Accuracy of the model on the 10000 test images: {accuracy:.2f}%')
```
